```python
import pandas as pd
import numpy as np

# Assuming 'molecules' DataFrame is already defined and contains 'pIC50' column

# Create a range of pIC50 values from 5 to 7 with a step of 0.1
pIC50_values = np.arange(5, 7.1, 0.1)  # 7.1 to include 7.0

# Create an empty list to store the results
results = []

# Iterate through the pIC50 values
for threshold in pIC50_values:
    # Count the number of active and inactive compounds
    active_count = (molecules['pIC50'] >= threshold).sum()  # Active if pIC50 >= threshold
    inactive_count = (molecules['pIC50'] < threshold).sum()  # Inactive if pIC50 < threshold
    
    # Append the results to the list
    results.append({'pIC50 Threshold': threshold, 'Active Count': active_count, 'Inactive Count': inactive_count})

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results)

# Display the resulting DataFrame
results_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample DataFrame creation (replace this with your actual DataFrame)
# molecules = pd.DataFrame({'pIC50': [2.0, 3.5, 5.5, 6.0, 6.5, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0]})

# Create a new column to categorize active and inactive
molecules['Activity'] = ['Active' if x >= 6.3 else 'Inactive' for x in molecules['pIC50']]

# Set the color palette
palette = {'Active': 'blue', 'Inactive': 'orange'}

# Define bins with a step of 0.1 from 2 to 12
bins = np.arange(2, 12.1, 0.1)  # 12.1 to include 12 in the last bin

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data=molecules, x='pIC50', bins=bins, kde=False, hue='Activity', palette=palette, edgecolor='k', stat='density')

# Add title and labels
plt.title('Distribucija pIC50 vrednosti')
plt.xlabel('pIC50')
plt.ylabel('Density')

# Manually create the legend with correct colors
handles = [
    plt.Line2D([0], [0], color=palette['Active'], lw=4, label='Aktivne molekule (pIC50 â‰¥ 6.3)'),
    plt.Line2D([0], [0], color=palette['Inactive'], lw=4, label='Neaktivne molekule (pIC50 < 6.3)')
]
plt.legend(handles=handles, title='Aktivnost molekul')

# Show grid
plt.grid()

# Display the plot
plt.show()
```
