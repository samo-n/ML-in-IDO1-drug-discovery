```python
import pandas as pd

# Define the fingerprint types
cdktypes = ['standard', 'extended', 'graph', 'maccs', 'pubchem', 'estate', 
            'hybridization', 'lingo', 'klekota-roth', 'shortestpath', 
            'cdk-substructure', 'circular', 'cdk-atompairs']
rdktypes = ['rdkit', 'morgan', 'rdk-maccs', 'topological-torsion', 'avalon', 'atom-pair']
babeltypes = ['fp2']

# Combine all fingerprint types
fingerprints = cdktypes + rdktypes + babeltypes

# Create a dictionary to hold DataFrames for each fingerprint type
fingerprint_dfs = {}

# Iterate through each fingerprint type and calculate fingerprints
for fingerprint in fingerprints:
    df_with_fingerprints, fingerprints_df = calc_fingerprints(df, fingerprint)
    fingerprint_dfs[f'df_{fingerprint}'] = df_with_fingerprints

# Define the directory to save the CSV files
output_directory = '/Users/samonose/PycharmProjects/IDO_inhibitors_ML_predictions/dir1/molekulski_prstni_odtisi'

# Save each DataFrame as a CSV file in the specified directory
for df_name, df in fingerprint_dfs.items():
    csv_file_name = f"{output_directory}/{df_name}.csv"  # Create a full path for the CSV file
    df.to_csv(csv_file_name, index=False)  # Save the DataFrame to a CSV file without the index
    print(f"Saved {df_name} as {csv_file_name}")
```
